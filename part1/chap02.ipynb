````python
!pip install lxml
!pip install bs4
````

````python
import pandas as pd

url = './data/sample.html'

tables = pd.read_html(url)

print(len(tables))

for i in range(len(tables)):
    print("tables[%s]" % i)
    print(tables[i])
    print("\n")

df = tables[1]
print(df)

df = df.set_index(['name'])
print(df)
````

# 데이터 저장하기
## csv 파일로 저장
```
CSV 파일로 저장: DataFrame 객체.to_csv("파일 이름(경로)")
```

````python
import pandas as pd

data = {'name' : ['Jerry', 'Riah', 'Paul'],
        'algo1' : ['A', 'A+', 'B'],
        'basic' : ['C', 'B', 'B+'],
        'c++' : ['B+', 'C', 'C+']
        }

df = pd.DataFrame(data)
df = df.set_index('name')

print(df)
````

- index=False 옵션은 데이터 프레임을 CSV로 저장할 때 행 인덱스를 파일에 포함하지 않도록 하고 데이터만을 포함한다.

````python
df.to_csv('./data/df_sample_no_index.csv', index=False)
````

## JSON 파일로 저장
```
JSON 파일로 저장 : DataFrame 객체.to_json("파일 이름(경로)")
```

````python
import pandas as pd

data = {'name' : ['Jerry', 'Riah', 'Paul'],
        'algo1' : ['A', 'A+', 'B'],
        'basic' : ['C', 'B', 'B+'],
        'c++' : ['B+', 'C', 'C+']
        }

df = pd.DataFrame(data)
df = df.set_index('name')

print(df)
````

````python
df.to_json("./data/df_sample.json")
````

orient="split" 옵션을 적용하면 데이터를 index, columns, data로 분리하여 저장한다.

````python
df.to_json("./data/df_sample_split.json", orient="split")
````

orient="records" 옵션을 적용하면 각 행을 별도의 JSON 객체로 저장하며, 열 이름이 키로 사용된다.

````python
df.to_json("./data/df_sample_records.json", orient="records")
````

- orient="index" 옵션을 적용하면 각 행의 인덱스를 키로 사용하고 각 행 데이터를 JSON 객체로 저장하여 매칭한다.

````python
df.to_json("./data/df_sample_index.json", orient="index")
````

- orient="column" 옵션을 적용하면 각 열을 별도의 JSON 객체로 저장하며, 열 이름이 키로 된다.

````python
df.to_json("./data/df_sample_columns.json", orient="columns")
````

- orient="values" 옵션을 적용하면 데이터프레임의 값만을 순서대로 작성한다.

````python
df.to_json("./data/df_sample_values.json", orient="values")
````

# 여러 개의 데이터프레임을 하나의 Excel 파일로 저장
- 판다스 ExcelWriter() 함수는 Excel 워크북 객체를 생성한다. 워크북 객체는 우리가 알고 있는 엑셀 파일이라고 생각하면 된다.
- 데이터 프레임에 to_excel() 메서드를 적용할 때 삽입하려는 워크북 객체를 인자로 전달한다.
- 또한 sheet_name 옵션에 Excel 파일의 시트 이름을 입력하여 삽입되는 시트 위치를 지정할 수 있다.
```
데이터프레임 여러 개를 Excel 파일로 저장 : pandas.ExcelWriter("파일 이름(경로)")
```

````python
import pandas as pd

data1 = {'name' : ['Jerry', 'Riah', 'Paul'],
         'algo1' : ['A', 'A+', 'B'],
         'basic' : ['C', 'B', 'B+'],
         'c++': ['B+', 'C', 'C+']
         }

data2 = {
    'c0':[1,2,3],
    'c1':[4,5,6],
    'c2':[7,8,9],
    'c3':[10,11,12],
    'c4':[13,14,15]
}

df1 = pd.DataFrame(data1)
df1 = df1.set_index('name')
print(df1)
print('\n')

df2 = pd.DataFrame(data2)
df2 = df2.set_index('c0')
print(df2)

with pd.ExcelWriter('./data/df_excel_sample.xlsx') as writer:
    df1.to_excel(writer, sheet_name='sheet1')
    df2.to_excel(writer, sheet_name='sheet2')
````
